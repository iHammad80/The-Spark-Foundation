# Decision Tree

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
df_modified = pd.read_csv('Iris.csv')
df_modified
```

```python
input1 = df_modified.drop(['Id','Species'], axis = 1)
target = df_modified['Species']
```

```python
model = DecisionTreeClassifier()
model.fit(input1, target)
```

#### Plotting Tree

```python
tree.plot_tree(model)
```

#### More clear Visuals

```python
feature = ['Sepal length (cm)', 'Sepal width (cm)','Petal length (cm)', 'Petal width (cm)']
Classes = ['Setosa', 'Versicolor', 'Verginica']
fig, axes = plt.subplots(nrows=1, ncols=1, figsize = (4,4),dpi = 300)

tree.plot_tree(model, feature_names=feature, class_names=Classes, filled=True)
```

#### Prediction for new test value.

```python
Sepal_length = input("Enter Sepal Length: ")
Sepal_width = input("Enter Sepal Width: ")
Petal_length = input("Enter Petal Length: ")
Petal_width = input("Enter Petal Width: ")
print(f'\nThe class according to the inputs, provided by user should be: {model.predict([[Sepal_length, Sepal_width, Petal_length, Petal_width]])}')
```
